```
#https://es.wikipedia.org/wiki/Conjunto_de_datos_flor_iris
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import sklearn
import pandas as pd
```

```
iris = sklearn.datasets.load_iris()
print(type(iris))
print(type(iris.data))
```

```
df=pd.DataFrame(iris.data, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])
```

```
df['species_id']=iris.target
```

```
df.head()
```

```
species_map = {0: 'setosa', 1: 'versicolor', 2:'virginica'}
```

```
df['species_name']=df['species_id'].map(species_map)
```

```
df.head()
```

```
from sklearn.model_selection import train_test_split

X=df[['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)']]

y=pd.DataFrame(df, columns=['species_id'])

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=12, test_size=0.20)#0.80 for training #shuffle : boolean, optional (default=True)
```

```
y_test
```

```
#USAR NORMALIZACION Y ESCALACION DE DATOS PARA VER SI INFLUYE EN NUESTRA PRUEBA
from sklearn import preprocessing
```

```
#ESCALADO
X_train_scaled = preprocessing.scale(X_train)
X_train_scaled=pd.DataFrame(X_train_scaled, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])

#To apply the same to test
scaler = preprocessing.StandardScaler().fit(X_train)
X_test_scaled=scaler.transform(X_test)  
X_test_scaled=pd.DataFrame(X_test_scaled, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])

X_train_scaled.describe()
```

```
#ESCALADO, deja el centro en 0 y de extremo a extremo hay 1
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
X_train_scaled_minmax = min_max_scaler.fit_transform(X_train)
X_test_scaled_minmax = min_max_scaler.transform(X_test)

X_train_scaled_minmax=pd.DataFrame(X_train_scaled_minmax, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])
X_test_scaled_minmax=pd.DataFrame(X_test_scaled_minmax, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])

X_train_scaled_minmax.describe()
```

```
#NORMALIZADO l1
from sklearn.preprocessing import Normalizer

normalizer = Normalizer(norm='l1').fit(X)
X_train_normalization_l1 = normalizer.transform(X_train)
X_test_normalization_l1 = normalizer.transform(X_test)

X_train_normalization_l1=pd.DataFrame(X_train_normalization_l1, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])
X_test_normalization_l1=pd.DataFrame(X_test_normalization_l1, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])

X_train_normalization_l1.describe()
```

```
#NORMALIZADO l2
from sklearn.preprocessing import Normalizer

normalizer = Normalizer(norm='l2').fit(X)
X_train_normalization_l2 = normalizer.transform(X_train)
X_test_normalization_l2 = normalizer.transform(X_test)

X_train_normalization_l2=pd.DataFrame(X_train_normalization_l2, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])
X_test_normalization_l2=pd.DataFrame(X_test_normalization_l2, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])

X_train_normalization_l2.describe()
```

```
#NORMALIZADO max
from sklearn.preprocessing import Normalizer

normalizer = Normalizer(norm='max').fit(X)
X_train_normalization_max = normalizer.transform(X_train)
X_test_normalization_max = normalizer.transform(X_test)

X_train_normalization_max=pd.DataFrame(X_train_normalization_max, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])
X_test_normalization_max=pd.DataFrame(X_test_normalization_max, columns=['sepal_length(cm)','sepal_width(cm)','pedal_length(cm)','pedal_width(cm)'])

X_train_normalization_max.describe()
```

```
#vamos a utilizar un analisis basado en : 
#https://machinelearningmastery.com/machine-learning-in-python-step-by-step/
#from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix #para evaluar
```

```
models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC(gamma='auto')))
# evaluate each model in turn

names = []
print("*********************************DATOS CRUDOS***************************")
for name, model in models:
  model.fit(X_train, np.ravel(y_train.values,order = 1))
  
  
  pred=model.predict(X_test)#predigo los resultados
  #pred=pd.DataFrame(pred, columns=['prediccion'])
  
  print("MODELO ELEGIDO: ",name," / Accuracy  :", sklearn.metrics.accuracy_score(y_test, pred))
  #print(confusion_matrix(y_test,pred))
  #print("--------------------------------------------------------------")
```
